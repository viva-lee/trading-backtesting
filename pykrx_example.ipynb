```python
from pykrx import stock
import pandas as pd
```

# 1. 종목별 ohlcv 값 가져오기

```python
# 2022년 8월 17일 코스피 종목별 ohlcv 값 가져오기
kospi_df = stock.get_market_ohlcv(date="20220817", market="KOSPI")
kospi_df
```

```python
kosdaq_df = stock.get_market_ohlcv(date="20220817", market="KOSDAQ")
kosdaq_df
```

# 2. 티커 값에 따른 종목명 확인

```python
stock.get_market_ticker_name("054620")
```

```python
kospi_df.reset_index(inplace=True)
kospi_df["종목명"] = kospi_df["티커"].apply(lambda x: stock.get_market_ticker_name(x))
```

```python
kospi_df
```

# 3. 등락률 기준 상위 10종목

```python
kospi_df = kospi_df.sort_values(by="등락률", ascending=False)[:10]
kospi_df
```

# 4. 등락률 10% 이상 종목 추출

```python
kospi_df = kospi_df[kospi_df["등락률"] > 10]
kospi_df
```
